```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.signal import gaussian
```

```python
def rgb2gray(x):
    return np.dot(x[:,:,:3],[0.2989,0.5870,0.1140])
def gauss(sz,std):
    x=np.matmul(gaussian(sz,std).reshape((-1,1)),gaussian(sz,std).reshape((1,-1)))
    return x
def sin_wav(sz,th,freq,ph):
    x=np.sin(np.matmul((np.arange(0,2*np.pi*freq,2*np.pi*freq/sz)*np.sin(th)).reshape((-1,1)),np.ones((1,sz)))-np.matmul(np.ones((sz,1)),(np.arange(0,2*np.pi*freq,2*np.pi*freq/sz)*np.cos(th)).reshape((1,-1)))+ph)
    return x
def gabor_wav(sz,std,th,freq,ph):
    return sin_wav(sz,th,freq,ph)*gauss(sz,std)
def four(x):
    return np.fft.fftshift(np.fft.fft2(x))
def four_st(x):
    return np.stack(np.fft.fftshift(np.fft.fft2(x[i,:,:])) for i in range(x.shape[0]))
def inv_four(x):
    return np.fft.ifft2(np.fft.ifftshift(x))
def gabor_f(sz,std,th,freq,ph):
    return four(gabor_wav(sz,std,th,freq,ph))
def four_viz(x):
    return np.log(np.maximum(np.abs(x),0.001))
#plt.imshow(gabor_wav(512,100,np.pi/4,4,0),cmap='gray')
#plt.show()
#plt.imshow(np.log(np.maximum(np.abs(gabor_f(512,100,np.pi/4,4,0)),0.001)),cmap='gray')
#plt.show()
```

```python
def gabor_calc(x,s):
    x=x.reshape(-1,1,s,s)
    freqs=np.arange(0,10,2)+2
    ors=np.arange(0,2*np.pi,np.pi/4)+np.pi/4
    f=[]
    #Permute set of filters
    for i in freqs:
        for j in ors:
            f.append(gabor_f(s,s/8,j,i,0))
    
    #Stack filters
    f=np.stack(f)
    
    #Calculate FFT
    x_f=four_st(x)
    
    #Convolve Gabor filter with frequency representation of images
    x_t=x_f*f
    
    #Sample 100 points for each image for each filter
    x_v=x_t[:,:,20::int(s/10),20::int(s/10)]
    
    #Compress into column vector
    x_c=np.stack([np.abs(x_v),np.angle(x_v)],axis=1).reshape(x_v.shape[0],-1)
    
    return x_c
```

```python
x=rgb2gray(plt.imread("1.png"))
y=rgb2gray(plt.imread("2.png"))
z=rgb2gray(plt.imread("3.png"))
plt.imshow(x,cmap='gray')
plt.show()
plt.imshow(y,cmap='gray')
plt.show()
plt.imshow(z,cmap='gray')
plt.show()
```

```python
stax=np.stack([x,y,z])
res=gabor_calc(stax,256)
```

```python
fig1,ax1=plt.subplots(1,2)
ax1[0].imshow(x,cmap='gray')
ax1[1].imshow(y,cmap='gray')
fig1.show()
fig1.suptitle("1 and 2: "+str(np.linalg.norm(res[0]-res[1])))

fig2,ax2=plt.subplots(1,2)
ax2[0].imshow(x,cmap='gray')
ax2[1].imshow(z,cmap='gray')
fig2.show()
fig2.suptitle("1 and 3: "+str(np.linalg.norm(res[0]-res[2])))

fig3,ax3=plt.subplots(1,2)
ax3[0].imshow(y,cmap='gray')
ax3[1].imshow(z,cmap='gray')
fig3.show()
fig3.suptitle("2 and 3: "+str(np.linalg.norm(res[1]-res[2])))
```
